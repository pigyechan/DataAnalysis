```python
import nltk
import konlpy
from konlpy.corpus import kobill

files_ko = kobill.fileids()
files_ko
```

```python
doc_ko = kobill.open('1809893.txt').read()
doc_ko
```

```python
from konlpy.tag import Okt
okt=Okt()

nouns = okt.nouns(doc_ko)
nouns
```

```python
ko=nltk.Text(nouns, name='의안 제1809893호')
print(len(ko.tokens))
print(len(set(ko.tokens)))
ko.vocab()
```

```python
from wordcloud import WordCloud
stopwords=['경우','만','액','세','제', '이', '의', '위','월','일', '관', '안','중','것', '호', '수', '항', '표','더', '항']
ko=[each_word for each_word in ko if each_word not in stopwords]
ko
```

```python
ko=nltk.Text(ko, name='의안 제1809893호')

plt.figure(figsize = (17,10))
ko.plot(50)
plt.show()
```

```python
data=ko.vocab().most_common(30)

wordcloud=WordCloud(font_path='/Library/Fonts/NanumBarunGothic.ttf',background_color="white").generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
